```python
import nltk
from nltk.corpus import stopwords
import re
```

```python
stop = set(stopwords.words("english"))
def clean_text(s):
    if pd.isna(s):
        return ""
    s = str(s).lower()
    s = re.sub(r"http\S+"," ", s)         # remove urls
    s = re.sub(r"@\w+"," ", s)            # remove mentions
    s = re.sub(r"[^a-z0-9\s]"," ", s)     # keep letters/numbers/spaces
    # Ø­Ø°Ù stopwords (Ø§Ø®ØªÛŒØ§Ø±ÛŒ ÙˆÙ„ÛŒ Ù…ÙÛŒØ¯Ù‡)
    words = [w for w in s.split() if w not in stop]
    return " ".join(words)
def tokenize(text):
  if pd.isna(text):
    return []
  return text.split()
```

```python
df_tweets['clean_text'] = df_tweets['text'].map(clean_text, meta=('text', 'str'))
texts = df_tweets['clean_text'].dropna().compute()
```

```python
unique_words = set()
for t in texts:
    unique_words.update(tokenize(t))
print(len(unique_words))
```
